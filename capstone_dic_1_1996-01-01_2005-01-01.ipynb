```python

# Load environment variables from the .env file
env_vars = dotenv_values(".env")

# Access the credentials using the loaded environment variables
api_key = env_vars["X-RapidAPI-Key"]
api_host = env_vars["X-RapidAPI-Host"]

# Use the credentials in your code
url = "https://meteostat.p.rapidapi.com/stations/daily"
weather_data_all = []

for station in weather_stations:
    querystring = {"station": station, "start": "1996-01-01", "end": "2005-01-01"}
    print(f'This is the weather for {station}')
    headers = {
        "X-RapidAPI-Key": api_key,  # Use the loaded API key
        "X-RapidAPI-Host": api_host  # Use the loaded API host
    }

    response = requests.get(url, headers=headers, params=querystring)
    weather_data_all.append(response.json())
    time.sleep(3)
```

```python
weather_capstone = weather_data_all     
weather_df_capstone = pd.DataFrame([])
for station, json_element in zip(weather_stations,weather_capstone): #zipping weather_stations & weather_katrina
    weather_norm = pd.json_normalize(json_element, sep="_",
                                    record_path="data",  
                                    record_prefix="weather_", errors='ignore')
    weather_norm['station_id'] = station

    weather_df_capstone = pd.concat([weather_df_capstone, weather_norm], ignore_index=True)
weather_df_capstone.info()
```

```python
weather_df_capstone['weather_station_name'] = weather_df_capstone['station_id'].map(weather_stations_dict)
```
